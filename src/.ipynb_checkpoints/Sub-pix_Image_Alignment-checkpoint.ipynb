```python
import numpy as np
import matplotlib.pyplot as plt
import hyperspy.api as hs
import os
```

```python
input_file = "Series_JEOL_1K_2usec.dm3"

cur_dir = os.getcwd()    # Get filepath of current directory
if os.name == "nt":      # If os is Windows
    data_dir = os.path.dirname(cur_dir) + "\\data"
    ds = hs.load(data_dir + "\\" + input_file)
if os.name == "posix":   # If os is Linux or Mac
    data_dir = os.path.dirname(cur_dir) + "//data"
    ds = hs.load(data_dir + "//" + input_file)
```

```python
s = hs.signals.Signal1D(ds)
zdim, ydim, xdim = s.data.shape
image = np.flipud(s.data)    # Flip vertically because dm4 has inverse definition of y-axis against matplotlib
image_resample = image.repeat(10, axis=1).repeat(10, axis=2)
```
